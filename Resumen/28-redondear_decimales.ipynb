### **28 - Redondear decimales**

```python
import pandas as pd
import numpy as np
```

#### **`>` round( )**

```python
df = pd.read_csv('flights.csv')
df
```

```python
df.groupby('year').agg({'passengers':'mean'})
```

```python
df.groupby('year').agg({'passengers':'mean'}).round()
```

```python
df.groupby('year').agg({'passengers':'mean'}).round(1)
```

```python
df.groupby('year').agg({'passengers':'mean'}).round(2)
```

#### **`>` astype('`int`')**

```python
df = pd.read_csv('flights.csv')
df
```

```python
df.groupby('year').agg({'passengers':'mean'}).astype('int')
```

#### **`>` np.round( )**

```python
df = pd.DataFrame(np.array(([1.35, 2.75, 3.22], [4.45, 5.66, 6.87])),
                  index=['mouse', 'rabbit'],
                  columns=['one', 'two', 'three'])
df
```

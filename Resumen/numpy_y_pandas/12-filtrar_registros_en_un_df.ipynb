#### **`>` loc( )**

Con el comando loc podemos obtener ciertas filas o columnas de un Dataframe a partir del nombre o etiqueta de los indices y columnas.En caso que no se encuentre el valor del indice o el nombre de la columna va entregar un error.

> Sintaxis: dataframe.loc[indices, columnas]

- **`Filas`**: corresponde al valor del indice del Dataframe, el cual podria ser String o Integer. Tambien es posible colocar condiciones que entreguen una lista de booleanos, donde filtra quedandose con los valores que son True.
- **`Columnas`**: Corresponde a los nombres de las columnas y tambiÃ©n se puede entregar una lista de booleanos

Solo Filas

```python
df.loc[1:2]
```

```python
# El segundo parametro corresponde a las columnas que necesitemos llamar. Si no colocamos nada, se llamaran a todas las columnas
# o simplemente tambiÃ©n podemos colocar :
df.loc[1:2, :]
```

Solo columnas

```python
# El primer parametro corresponde a las filas que necesitemos llamar. En este caso, si es obligatorio colocar : en el primer
# argumento al momento de filtrar columnas:
df.loc[:, ['team', 'continent']].head()
```

Filas y columnas

```python
df.loc[0:3, ['team', 'continent']]
```

Filtrar con condiciones filas

```python
df.loc[df.continent=='asia']
```

```python
df.loc[df.continent=='asia', 'continent']
```

```python
df.loc[df.continent=='asia', ['team','continent','group']]
```

Filtrar con condiciones columna

```python
df.columns.str.contains('juegos')
```

```python
df.loc[:, df.columns.str.contains('juegos')].head()
```

Filtrar con condiciones filas y condiciones columnas

```python
df.loc[df.continent=='asia', df.columns.str.contains('juegos')]
```

#### **`>` iloc( )**

Con el comando iloc podemos obtener ciertas filas o columnas de un Dataframe a partir de la posiciÃ³n que se encuentran los indices y columnas.En caso que el valor consultado este fuera del rango entrega error.

> Sintaxis: dataframe.iloc[indices, columnas]

- **`Filas`**: corresponde a la posiciÃ³n de la fila que se quiere obtener, solo puede ser numerico
- **`Columnas`**: Corresponde a la posicion de la columna que se desea obtener

Solo Filas

```python
# Tomara la solo la fila 1
df.iloc[1:2]
```

```python
# El segundo parametro corresponde a las columnas que necesitemos llamar. Si no colocamos nada, se llamaran a todas las columnas
# o simplemente tambiÃ©n podemos colocar :
df.iloc[1:2, :]
```

```python
# Tomara la solo las filas 1 y 2
df.iloc[1:3]
```

Solo columnas

```python
# El primer parametro corresponde a las filas que necesitemos llamar. En este caso, si es obligatorio colocar : en el primer
# argumento al momento de filtrar columnas:
df.iloc[:, 1:5].head()
```

Filas y columnas

```python
df.iloc[1:4, 1:5]
```

#### **`>` where( )**

- El mÃ©todo **`where()`** de Pandas se utiliza para comprobar un dataframe para una o mÃ¡s condiciones y devolver el resultado en consecuencia. Por defecto, las filas que no satisfacen la condiciÃ³n se rellenan con un valor **`NaN`**.

**`Sintaxis`**: DataFrame.where(
    cond,
    other=NaN,
    inplace=False,
    axis=None,
    level=None,
    errors="raise",
    try_cast=False
)

- **`condition`**:	Es una Series o DataFrame booleana, una estructura de tipo array o, una llamada. Representa la condiciÃ³n/condiciones para comprobar cada valor del DataFrame. Si la condiciÃ³n es True, entonces el valor original no es reemplazado. De lo contrario, es reemplazado por un valor **`NaN`**.
- **`other`**:	Es un escalar, Series/DataFrame, o un llamable. Representa el valor que se colocarÃ¡ para el valor original si la condiciÃ³n es False.
- **`inplace`**:	Es un valor booleano. Habla de la operaciÃ³n sobre los datos. Si es True, hace los cambios por sÃ­ mismo.
- **`axis`**:	Es un valor entero. Indica el eje de trabajo, ya sea filas o columnas.
- **`level`**:	Es un valor entero. Dice sobre el nivel.
- **`errors`**:	Es una cuerda. Cuenta los errores. Acepta dos opciones: raise o ignore. Si su valor es raise entonces permite que las excepciones sean aumentadas. Si su valor es ignore, entonces ignora las excepciones y devuelve el objeto original si hay un error.
- **`try_cast`**:	Es un valor booleano. Echa la salida de la funciÃ³n al tipo de entrada original si es posible.

##### Ejemplo 1

```python
de = pd.DataFrame(dict(age=[5, 6, np.nan],
                       born=[pd.NaT, pd.Timestamp('1939-05-27'),
                             pd.Timestamp('1940-04-25')],
                       name=['Alfred', 'Batman', ''],
                       toy=[None, 'Batmobile', 'Joker']))

de
```

```python
de.where(de['toy']=='Joker')
```

```python
de.where(de['toy']=='Joker').dropna(how='all')
```

##### Ejemplo 2

```python
data = pd.read_csv("nba.csv") 
data
```

```python
# Forma 1
dp = data.sort_values("Team") 

filtro = data["Team"]=="Atlanta Hawks"

# Filtrando la data
dp.where(filtro) 
```

```python
# Forma 2
data.sort_values("Team").where(data["Team"]=="Atlanta Hawks")
```

```python
data.sort_values("Team").where(data["Team"]=="Atlanta Hawks").dropna(how='all')
```

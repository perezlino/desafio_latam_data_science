### **2 - Matplotlib.pyplot.subplot( )**

#### **`Matplotlib.pyplot.subplot()`**

La función **`subplot()`** añade una subplot a una figura actual en la posición de cuadrícula especificada.  Es similar a la función **`subplots()`** sin embargo, a diferencia de **`subplots()`**, añade un subplot cada vez. Por lo tanto, para crear múltiples gráficos necesitará varias líneas de código con la función **`subplot()`**. Otro inconveniente de la función subplot es que borra el gráfico preexistente en su figura. Consulte el ejemplo 1.

Es un wrapper de **`Figure.add_subplot`**.

#### **`Sintaxis de Matplotlib pyplot.subplot()`**

**`Sintaxis`**:

- subplot(nrows, ncols, index, **kwargs)
- subplot(pos, **kwargs) 
- subplot(ax)

**`Parámetros`** : 

- **`args`**:  Un entero de 3 dígitos o tres enteros separados que describen la posición del subplot.
- **`pos`** es un entero de tres dígitos donde el primero, segundo y tercer entero son nrows, ncols, index.
- **`projection`** : [{Ninguna, 'aitoﬀ', 'hammer', 'lambert', 'mollweide', 'polar', 'rectilinear', str}, opcional]. Tipo de proyección de la subparcela (Axes). El valor por defecto None da como resultado una proyección 'rectilínea'.
- **`label`** : [str] Una etiqueta para los ejes devueltos.
- **`**kwargs`**: Este método también toma los argumentos keyword para la clase base de los ejes devueltos; excepto el argumento ﬁgure, por ejemplo facecolor.

**`Devuelve`** : Una subclase axes.SubplotBase de Axes o una subclase de Axes. La clase base de ejes devuelta depende de la proyección utilizada.

#### **Ejemplo 2**

```python
import matplotlib.pyplot as plt

# datos a mostrar en los gráficos
x = [3, 1, 3]
y = [3, 2, 1]
z = [1, 3, 1]

# Crear objeto figura
plt.figure()

# añadir primer subplot
# num debe ser 1 <= num <= 2
plt.subplot(121)
plt.plot(x, y, color="orange", marker="*")

# añadir segundo subplot
# num debe ser 1 <= num <= 2
plt.subplot(122)
plt.plot(z, y, color="yellow", marker="*")
```

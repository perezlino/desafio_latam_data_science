### **5 - Matplotlib.pyplot.ylabel( )**

#### **`Matplotlib.pyplot.ylabel()`**

Esta función establece la etiqueta para el eje y del gráfico.

#### **`Sintaxis de Matplotlib pyplot.ylabel()`**

**`Sintaxis`**: 

> **matplotlib.pyplot.ylabel**(**ylabel**, **fontdict**=None, **labelpad**=None)

**`Parámetros`**: Este método acepta los siguientes parámetros que se describen a continuación:

- **`ylabel`**: El nombre de la etiqueta

- **`fontdict`**: Añade los estilos de fuente a la etiqueta

- **`labelpad`**: Nos ayuda a establecer el espaciado entre la etiqueta y el eje

##### **Ejemplo 1**

```python
import matplotlib.pyplot as plt 

# establecer valores x
x =['Geeks', 'for', 'geeks', 'tutorials'] 

# Ajuste de los valores y
y =[1, 2, 3, 4] 

# Añadir etiqueta en el eje y
plt.ylabel('Numbers label') 

# trazar el gráfico
plt.plot(x, y) 
```

##### **Ejemplo 2**

```python
import matplotlib.pyplot as plt 

x =['Geeks', 'for', 'geeks', 'tutorials'] 
y =[1, 2, 3, 4] 

# Añadir espacio entre etiqueta y eje ajustando labelpad 
plt.ylabel('Numbers label', labelpad = 100) 

plt.plot(x, y) 
```

##### **Ejemplo 3**

```python
import matplotlib.pyplot as plt 

x =['Geeks', 'for', 'geeks', 'tutorials'] 
y =[1, 2, 3, 4] 

# Establecer diccionario de fuentes
font = {'family': 'Verdana', 
		'color': 'green', 
		'size': 20, 
		} 

# Añadir los estilos de fuente a la etiqueta
plt.ylabel('Numbers label', fontdict = font) 

plt.plot(x, y) 
```

##### **Ejemplo 4**

```python
import numpy as np 
import matplotlib.pyplot as plt 

x2 = np.linspace(0.0, 2.0) 
y2 = np.cos(2 * np.pi * x2) 

plt.plot(x2, y2, '.-')
plt.title('xlabel() Example') 
plt.xlabel('time (s)') 
plt.ylabel('Undamped', fontsize=20) 
plt.show()
```

### **4 - Matplotlib.pyplot.xlabel( )**

#### **`Matplotlib.pyplot.xlabel()`**

La función **`xlabel()`** en el módulo pyplot de la biblioteca matplotlib se utiliza para establecer la etiqueta para el eje x..

#### **`Sintaxis de Matplotlib pyplot.xlabel()`**

**`Sintaxis`**: 

> **matplotlib.pyplot.xlabel**(**xlabel**, **fontdict**=None, **labelpad**=None, ****kwargs**)

**`Parámetros`**: Este método acepta los siguientes parámetros que se describen a continuación:

- **`xlabel`**: Este parámetro es el texto de la etiqueta. Y contiene el valor del string.

- **`labelpad`**: Este parámetro se utiliza para el espaciado en puntos de la caja delimitadora de los ejes incluyendo ticks y tick labels y su valor por defecto es None.

- **`**kwargs`**: Este parámetro es una propiedad de texto que se utiliza para controlar la apariencia de las etiquetas.

#### **Ejemplo 1**

```python
import numpy as np 
import matplotlib.pyplot as plt 

t = np.arange(-180.0, 180.0, 0.1) 
s = np.radians(t)/2.

plt.plot(t, s, '-', lw = 2) 

plt.xlabel('Longitude') 
plt.ylabel('Latitude') 
plt.title('xlabels() function') 
plt.grid(True) 

plt.show() 
```

#### **Ejemplo 2**

```python
import numpy as np 
import matplotlib.pyplot as plt 

valx1 = np.linspace(0.0, 5.0) 
x2 = np.linspace(0.0, 2.0) 

valy1 = np.cos(2 * np.pi * valx1) * np.exp(-valx1) 
y2 = np.cos(2 * np.pi * x2) 

# Establecer diccionario de fuentes
font = {'family': 'Verdana', 
        'color':  'green', 
        'size': 20, 
        } 

plt.subplot(2, 1, 1) 
plt.plot(valx1, valy1, 'o-') 
plt.title('xlabel() Example') 
plt.ylabel('Damped oscillation') 

plt.subplot(2, 1, 2) 
plt.plot(x2, y2, '.-')

# Añadir los estilos de fuente a la etiqueta 
plt.xlabel('time (s)', fontdict = font) 

plt.ylabel('Undamped') 

plt.show() 
```

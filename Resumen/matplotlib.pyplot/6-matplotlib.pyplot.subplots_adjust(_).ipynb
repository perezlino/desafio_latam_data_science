### **6 - Matplotlib.pyplot.subplots_adjust( )**

#### **`Matplotlib.pyplot.subplots_adjust()`**

La función subplots_adjust() del módulo pyplot de la biblioteca matplotlib se utiliza para ajustar el diseño de los subplots.

#### **`Sintaxis de Matplotlib pyplot.subplots_adjust()`**

**`Sintaxis`**: 

> **matplotlib.pyplot.subplots_adjust**(**left**=None, **bottom**=None, **right**=None, **top**=None, **wspace**=None, **hspace**=None)

**`Parámetros`**: Este método acepta los siguientes parámetros que se describen a continuación:

- **`left`** : Este parámetro es el lado izquierdo de los subplots de la figura.
- **`right`** : Este parámetro es el lado derecho de los subplots de la figura.
- **`bottom`** : Este parámetro es la parte inferior de los subplots de la figura.
- **`top`** : Este parámetro es la parte superior de los subplots de la figura.
- **`wspace`** : Este parámetro es la cantidad de anchura reservada para el espacio entre los subplots, expresada como fracción de la anchura media del eje.
- **`hspace`** : Este parámetro es la cantidad de altura reservada para el espacio entre los subplots, expresada como fracción de la altura media del eje.

#### **Ejemplo 1**

```python
import matplotlib.pyplot as plt 

x = [1, 12, 3, 9] 
y = [1, 4, 9, 16] 
labels = ['Geeks1', 'Geeks2', 'Geeks3', 'Geeks4'] 

plt.plot(x, y) 
plt.xticks(x, labels, rotation ='vertical') 

plt.margins(0.2) 
plt.subplots_adjust(bottom = 0.8) 

plt.title('matplotlib.pyplot.subplots_adjust() Example') 
plt.show() 
```

### **1 - Matplotlib.pyplot.plot( )**

#### **`Matplotlib.pyplot.plot()`**

La función **`matplotlib.pyplot.plot()`** es un componente fundamental de la librería Matplotlib, concretamente dentro del módulo Pyplot. Sirve para generar un gráfico hexagonal 2D basado en los puntos de datos dados representados por las variables x e y. Conecta los puntos de datos con líneas, permitiendo la personalización de la apariencia del gráfico a través de parámetros como estilos de línea y marcadores. Esta versátil función se utiliza ampliamente para la visualización de datos en diversos ámbitos.

Hay varias maneras de crear un gráfico usando la función Matplotlib.pyplot.plot() en Python, Estos son algunos ejemplos que ilustran la función **`matplotlib.pyplot.plot()`** en matplotlib.pyplot:

- Gráfico de líneas básico
- Gráfico de líneas múltiples
- Gráfico de dispersión con múltiples marcadores
- Gráfico de dos curvas

#### **`Sintaxis de Matplotlib pyplot.plot()`**

> **matplotlib.pyplot.plot**(***args**, **scalex**=True, **scaley**=True, **data**=None, ****kwargs**)

**`Parámetros`**:

- **`x, y`**: Estos parámetros representan las coordenadas horizontales y verticales de los puntos de datos. Los valores de 'x' son opcionales, lo que permite flexibilidad en el proceso de trazado.

- **`fmt`**: Es un parámetro opcional que contiene un valor de tipo string. Se utiliza para especificar el formato del gráfico, definiendo el estilo de línea, el marcador y el color.

- **`data`**: Un parámetro opcional, 'data' se refiere a un objeto con datos etiquetados. Proporciona una forma conveniente de pasar datos directamente, mejorando la legibilidad y facilidad de uso.

**`Devuelve`**: La función plot() devuelve una lista de objetos Line2D, cada uno de los cuales representa un segmento de los datos trazados. Estos objetos Line2D encapsulan las características y atributos de las líneas trazadas, permitiendo una mayor personalización y análisis.

#### **Ejemplo 1**

Al importar la función **`plot()`** de Matplotlib, hemos creado un gráfico de líneas con los datos [1, 2, 3]. La función **`title()`** establece el título del gráfico, **`draw()`** actualiza el gráfico y **`show()`** lo muestra, proporcionando una ilustración básica de Matplotlib para la visualización de datos en Python.

```python
import matplotlib.pyplot as plt
import numpy as np

# Plotting a simple line graph
plt.plot([1, 2, 3])

# Setting the title
plt.title('Matplotlib Line Plot Example')

plt.show()
```

#### **Ejemplo 2**

**`Múltiples líneas usando Matplotlib`**

Importando Matplotlib para trazar funciones seno y coseno en el mismo gráfico. Genera datos, establece estilos para cada función, añade etiquetas y un título, muestra una leyenda y, a continuación, muestra el gráfico, ilustrando las curvas seno y coseno.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos
x = np.linspace(0, 2*np.pi, 100)
y1, y2 = np.sin(x), np.cos(x)

# Trazado de múltiples líneas en un único gráfico
plt.plot(x, y1, label='Sin(x)', color='b')
plt.plot(x, y2, label='Cos(x)', color='r', linestyle='--')

# Añadir etiquetas y título
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines Plot')

# Visualización de la leyenda y del gráfico
plt.legend()
plt.show()
```

#### **Ejemplo 3**

**`Marcadores en Matplotlib`**

Importando Matplotlib generamos un gráfico de dispersión personalizado con 50 puntos de datos aleatorios, con marcadores circulares rojos. Incluye etiquetas de eje, un título ('Ejemplo de gráfico de dispersión') y una leyenda. La función show() muestra el gráfico, mostrando un ejemplo básico de visualización de datos con Matplotlib en Python.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos aleatorios
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

# Trazar un gráfico de dispersión con marcadores personalizados
plt.plot(x, y, marker='x', linestyle='', markersize=8, color='r', label='Scatter Plot')

# Añadir etiquetas y título
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Visualización de la leyenda
plt.legend()

# Visualizar el gráfico
plt.show()
```

#### **Ejemplo 4**

**`Trazado de múltiples curvas`**

Importando Matplotlib creamos un gráfico de líneas con dos curvas: una curva azul (y=x^2) y una curva naranja (y=1-x^3). Los datos se generan aleatoriamente, se ordenan para obtener curvas suaves y se trazan con la función plot(). El gráfico se limita al intervalo [0, 1] en ambos ejes, mostrando una representación visual de las funciones matemáticas.

```python
# Implementation of matplotlib function 

import matplotlib.pyplot as plt 
import numpy as np 

# Fijar el estado aleatorio para la reproducibilidad
# Siempre obtendremos el mismo conjunto de números aleatorios para la misma semilla en cualquier máquina
np.random.seed(19680801) 

# crear datos aleatorios
# xdata = np.random.random_sample([2, 10]) --> Lo mismo
# xdata = np.random.rand(2, 10) --> Lo mismo
xdata = np.random.random((2, 10)) # [[0.7003673  0.74275081 0.70928001 0.56674552 0.97778533 0.70633485 0.24791576 0.15788335 0.69769852 0.71995667]
                                  #  [0.25774443 0.34154678 0.96876117 0.6945071  0.46638326 0.7028127  0.51178587 0.92874137 0.7397693  0.62243903]]

# divide los datos en dos partes
xdata1 = xdata[0, :] # [0.7003673  0.74275081 0.70928001 0.56674552 0.97778533 0.70633485 0.24791576 0.15788335 0.69769852 0.71995667]
xdata2 = xdata[1, :] # [0.25774443 0.34154678 0.96876117 0.6945071  0.46638326 0.7028127  0.51178587 0.92874137 0.7397693  0.62243903]

# ordena los datos para que formen curvas limpias
xdata1.sort() 
xdata2.sort() 

# crear algunos puntos de datos y
ydata1 = xdata1 ** 2
ydata2 = 1 - xdata2 ** 3

# trazar los datos
plt.plot(xdata1, ydata1, color ='tab:blue') 
plt.plot(xdata2, ydata2, color ='tab:orange') 


# fijar los límites
plt.xlim([0, 1]) 
plt.ylim([0, 1]) 

plt.title('matplotlib.pyplot.plot() example 2') 

# mostrar el gráfico
plt.show()
```

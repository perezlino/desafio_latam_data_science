### **9 - Matplotlib.pyplot.margins( )**

#### **`Matplotlib.pyplot.margins()`**

**`Matplotlib.pyplot.margins()`** es una función utilizada para establecer los márgenes de los ejes x e y. Todos los parámetros de entrada deben ser un float que también dentro del rango [0, 1]. No podemos pasar argumentos posicionales y keyword a la vez ya que se producirá un TypeError. El padding añadido a cada límite de los ejes es el margen multiplicado por el intervalo de datos para ese eje. Si no se proporcionan argumentos, se mantienen los márgenes existentes. Especificar márgenes cambia el autoescalado. 

- Si sólo se proporciona un valor flotante, éste se toma como margen para los ejes x e y. Si se proporcionan dos valores flotantes, se toma como margen para el eje x.
- Si se proporcionan dos valores float, se tomarán para especificar los márgenes de los ejes x e y respectivamente.

#### **`Sintaxis de Matplotlib pyplot.margins()`**

**`Sintaxis`**:

> **margins**(**x**, **y**, **tight**)

**`Parámetros`**
- **`x, y`**: Permite especificar los valores de los márgenes para los ejes x e y. Sólo pueden utilizarse individualmente.
- **`tight`**: Booleano 
- Si este parámetro se especifica como True, se considera que los márgenes especificados no requieren ningún relleno adicional para ajustarse a las ticks.
- Si este parámetro se especifica como None, se conservará la configuración original.

#### **Ejemplo 1**

```python
import matplotlib.pyplot as plt 

x = [1, 2, 3, 4, 5] 
labels = ['Radius 1', 'Radius 2', 'Radius 3', 'Radius 4', 'Radius 5'] 
y = [i**2 for i in x] 

plt.plot(x, y) 
plt.xticks(x, labels, rotation=45) 
plt.margins() 
plt.show() 
```

##### **`iteritems()`**

- La dirección de recorrido es hacia la **`DERECHA`**
- Recorro columna por columna el Dataframe
- Me permite recorrer los nombres de las columnas

```python
for colname, serie in df.iteritems():
    print(colname)
```

- Me permite recorrer los datos de cada columna, columna por columna
- Primero la columna de AÑO, luego la columna de MES y finalmente la columna de PASAJEROS

```python
for colname, serie in df.iteritems():
    print(serie)
```

```python
for colname, serie in df.iteritems():
    print(serie.dtype)
```

##### **`iterrows()`**

- La dirección de recorrido es hacia la **`ABAJO`**
- Recorro fila por fila el Dataframe
- Me permite recorrer los indices de las filas

#### Ejemplo 1

```python
for idx, row in df.iterrows():
    print(idx)
```

#### Ejemplo 2

Podemos recorrer fila por fila

```python
for idx, row in df.iterrows():
    print(row)
    print()
```

#### Ejemplo 3

```python
for idx, row in dp.iterrows():
    print(row['nombre'])
    print()
```

#### Ejemplo 4

```python
for idx, row in dp.iterrows():
    print(idx, row[4])
    print()
```

#### Ejemplo 4

Iterar siempre es una forma ineficiente, imaginandonos que fueran a existir miles y miles de registros

```python
mean_male = 0
for index, row in dp.iterrows():
    if row['sexo'] == 'M':
        mean_male += row['peso']

print(mean_male/6)
```

Por tanto, podemos hacer lo siguiente:

```python
dp[dp['sexo'] == 'M']['peso'].mean()
```

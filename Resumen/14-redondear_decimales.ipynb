### **14 - Redondear decimales**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('flights.csv')
df
```

#### **`>` round()**

```python
df.groupby('year').agg({'passengers':'mean'})
```

```python
df.groupby('year').agg({'passengers':'mean'}).round()
```

```python
df.groupby('year').agg({'passengers':'mean'}).round(1)
```

```python
df.groupby('year').agg({'passengers':'mean'}).round(2)
```

#### **`>` astype('`int`')**

```python
df.groupby('year').agg({'passengers':'mean'}).astype('int')
```

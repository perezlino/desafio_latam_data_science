#### **`Recorrer un Dataframe`**

##### **`iteritems()`**

- La direcciÃ³n de recorrido es hacia la **`DERECHA`**
- Recorro columna por columna el Dataframe
- Me permite recorrer los nombres de las columnas

```python
for colname, serie in dt.iteritems():
    print(colname)
```

- Me permite recorrer los datos de cada columna, columna por columna
- Primero la columna de AÃ‘O, luego la columna de MES y finalmente la columna de PASAJEROS

```python
for colname, serie in dt.iteritems():
    print(serie)
```

Podemos tambiÃ©n ver los tipos de datos de cada una de estas columnas

```python
for colname, serie in dt.iteritems():
    print(serie.dtype)
```

##### **`iterrows()`**

- La direcciÃ³n de recorrido es hacia la **`ABAJO`**
- Recorro fila por fila el Dataframe
- Me permite recorrer los indices de las filas

```python
for idx, row in dt.iterrows():
    print(idx)
```

Podemos recorrer fila por fila

```python
for idx, row in dt.iterrows():
    print(row)
    print()
```

```python
for idx, row in df.iterrows():
    print(row['nombre'])
    print()
```

#### **`A nivel de Dataframe: iterrows condicionales`**

##### **`Ejemplo 1`**

Iterar siempre es una forma ineficiente, imaginandonos que fueran a existir miles y miles de registros

```python
mean_male = 0
for index, row in df.iterrows():
    if row['sexo'] == 'M':
        mean_male += row['peso']

print(mean_male/6)
```

Por tanto, podemos hacer lo siguiente:

```python
df[df['sexo'] == 'M']['peso'].mean()
```

```python
for index, row in df.iterrows():
    if row['sexo'] == 'M':
        print(row['peso'])
```

##### **`Ejemplo 2`**

Calcular el **`PROMEDIO DEL PESO`** segÃºn el **`SEXO`**

```python
s = df['sexo'].unique()

s
```

```python
for sexo in df['sexo'].unique():
    print(sexo)
```

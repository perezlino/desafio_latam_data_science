```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df  = pd.read_csv('qog_std_cs_jan18.csv')
df.head()
print(df.shape)
```

```python
hdi_drop = df['undp_hdi'].dropna()
plt.hist(hdi_drop, alpha = 0.5, color = "gray")
plt.grid()
plt.axvline(hdi_drop.mean(), color='tomato')
plt.axvline(hdi_drop.median(), color='black')
plt.show()
```

```python
group_hdi_mean = df.groupby('ht_region')['undp_hdi'].mean().sort_values()
plt.scatter(group_hdi_mean.values, group_hdi_mean.index, color = 'gray')
plt.title("Posición de distintas zonas geográficas en cuanto al HDI")
plt.grid()
plt.ylabel("Continentes")
plt.xlabel("Índice de desarrollo humano (HDI)")
plt.axvline(group_hdi_mean.mean(), color = 'orange', linestyle = '--')
```

```python
mu = hdi_drop.mean()
sigma = np.var(hdi_drop)

print(mu,sigma)
```

```python
simulate_normal = np.random.normal(mu,np.sqrt(sigma),1000)
plt.hist(simulate_normal, color = 'lightgray',alpha = 0.4, density = True)
plt.xlim(0.2, 1.2)
```

```python
from scipy.stats import norm
x_axis = np.linspace(0.2, 1.2, 1000)
simulate_normal_pdf = norm.pdf(x_axis, mu, np.sqrt(sigma))
plt.scatter(x_axis, simulate_normal_pdf, color = 'tomato', lw = 3)
```

```python
plt.scatter(x_axis, simulate_normal_pdf, color = 'tomato', lw = 3)
plt.hist(simulate_normal, color = 'lightgray',alpha = 0.4, density = True)
plt.xlim(0.2, 1.2)
```

```python
def get_zscore(col, case):
  get_point = df[df['cname'] == case][col]
  preproc = df[col].dropna()
  get_mean = preproc.mean()
  get_std = preproc.std()
  return float((get_point- get_mean)/get_std)
```

```python
get_zscore('undp_hdi','Chile')
```

#### **`>` loc( )**

Con el comando loc podemos obtener ciertas filas o columnas de un Dataframe a partir del nombre o etiqueta de los indices y columnas.En caso que no se encuentre el valor del indice o el nombre de la columna va entregar un error.

> Sintaxis: dataframe.loc[indices, columnas]

- **`Filas`**: corresponde al valor del indice del Dataframe, el cual podria ser String o Integer. Tambien es posible colocar condiciones que entreguen una lista de booleanos, donde filtra quedandose con los valores que son True.
- **`Columnas`**: Corresponde a los nombres de las columnas y también se puede entregar una lista de booleanos


Solo Filas

```python
df.loc[1:2]
```

```python
# El segundo parametro corresponde a las columnas que necesitemos llamar. Si no colocamos nada, se llamaran a todas las columnas
# o simplemente también podemos colocar :
df.loc[1:2, :]
```

Solo columnas

```python
# El primer parametro corresponde a las filas que necesitemos llamar. En este caso, si es obligatorio colocar : en el primer
# argumento al momento de filtrar columnas:
df.loc[:, ['team', 'continent']].head()
```

Filas y columnas

```python
df.loc[0:3, ['team', 'continent']]
```

Filtrar con condiciones filas

```python
df.loc[df.continent=='asia']
```

Filtrar con condiciones columna

```python
df.columns.str.contains('juegos')
```

```python
df.loc[:, df.columns.str.contains('juegos')].head()
```

Filtrar con condiciones filas y condiciones columnas

```python
df.loc[df.continent=='asia', df.columns.str.contains('juegos')]
```

#### **`>` iloc( )**

Con el comando iloc podemos obtener ciertas filas o columnas de un Dataframe a partir de la posición que se encuentran los indices y columnas.En caso que el valor consultado este fuera del rango entrega error.

> Sintaxis: dataframe.iloc[indices, columnas]

- **`Filas`**: corresponde a la posición de la fila que se quiere obtener, solo puede ser numerico
- **`Columnas`**: Corresponde a la posicion de la columna que se desea obtener


Solo Filas

```python
# Tomara la solo la fila 1
df.iloc[1:2]
```

```python
# El segundo parametro corresponde a las columnas que necesitemos llamar. Si no colocamos nada, se llamaran a todas las columnas
# o simplemente también podemos colocar :
df.iloc[1:2, :]
```

```python
# Tomara la solo las filas 1 y 2
df.iloc[1:3]
```

Solo columnas

```python
# El primer parametro corresponde a las filas que necesitemos llamar. En este caso, si es obligatorio colocar : en el primer
# argumento al momento de filtrar columnas:
df.iloc[:, 1:5].head()
```

Filas y columnas

```python
df.iloc[1:4, 1:5]
```
